```python
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
def quantile(a, qs = [0, 25, 50, 75, 100], dec=1):
    value = np.percentile(a, q=qs)
    df = pd.DataFrame({'quantile(%)': qs, 'value': np.around(value, decimals=dec)})
    return df
# end
```

```python
HOME_DIR = 'd:/kaggle/click_pred/'
doc_topic_pairs = pd.read_csv(os.path.join(HOME_DIR,"in/documents_topics.csv"))
```

```python
doc_categ = pd.read_csv(os.path.join(HOME_DIR, 'in/documents_categories.csv'))
```

```python
print 'Quartiles of topic conf. level:' 
print quantile(a=doc_topic_pairs['confidence_level'], dec=3)

print 'Quartiles of categ conf. level:'
print quantile(doc_categ['confidence_level'], dec=3)
```

```python
# Plot distributions of conf. level

fig = plt.figure()
fig.suptitle('Conf. level (denoted as c) of topics and categories in documents', fontsize=16, y=1.1)

# topic conf. level
plt.subplot(1, 2, 1)
n, bins, patches = plt.hist(doc_topic_pairs['confidence_level'], histtype='bar', log=True, color='g')
# print 'most frequent conf. level: %f' %bins[np.argmax(n)], ', with count: %d' %max(n)
# print 'least frequent conf. level: %f' %bins[np.argmin(n)], ', with count: %d' %min(n)
plt.title('Conf. level of topics', fontsize=12, )
plt.xlabel('c')
plt.ylabel('no. of topics')
plt.xlim(0, 1)

plt.subplots_adjust(wspace=1, hspace=1)

# categ conf. level
plt.subplot(1, 2, 2)
n, bins, patches = plt.hist(doc_categ['confidence_level'], histtype='bar', log=True)
plt.title('Conf. level of categories', fontsize=12)
plt.xlabel('c')
plt.ylabel('no. of categories')
plt.xlim(0, 1)

plt.show()
```
